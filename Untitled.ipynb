```python
#importing the libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict
```

```python
def plot_country(df, filename): 
    plt.figure(figsize=(20,10))
    ax = df['Country'].value_counts()[:20].plot(kind = 'bar', width = 0.8, color = '#9c9054')
    for p in ax.patches:
        ax.annotate(format(p.get_height()),(p.get_x()+0.15, p.get_height()+10))
    plt.xlabel('Country')
    plt.ylabel('Count')
    plt.title('Number of Responders per country (Top 20)')
    plt.savefig(f'./blog/{filename}.png')
    return plt.show()
    
plot_country(df, 'countywise')
```

```python
student = df[df['Professional']=='Student']
plot_country(student,'studcount')
```

```python
languages = {}
sd = student['HaveWorkedLanguage'].dropna()
for i in sd.values:
    for l in i.split(';'):
        languages[l.strip()] = languages[l.strip()] + 1 if l.strip() in languages else 1
```

```python
# print(f"There are {len(languages)} Languages")
def show_languages(lang,filename):
    lang = {k: v for k, v in sorted(lang.items(), key=lambda item: item[1])}
    plt.figure(figsize=(15,15))
    ax = pd.Series(lang).plot(kind='barh', width = 0.8, color = '#9c9054')
    for i,v in enumerate(lang.values()):
        ax.text(v + 3, i - 0.2, str(v), color='black')
    plt.xlabel("Count")
    plt.ylabel("Languages")
    plt.title("Number of students v/s Language")
    plt.savefig(f'./blog/{filename}.png')
    return plt.show()
show_languages(languages,'studentlang')
```

```python
print("Average Salary per formal education")
```

```python
def plot_formaleducation(df, column):
    plt.figure(figsize=(20,10))
    dt = df.dropna(subset=[column]).groupby('FormalEducation').mean()[column]
    ax = dt.plot(kind = 'barh', width = 0.8, color = '#9c9054')
    for i,v in enumerate(dt):
        ax.text(v + v/200, i, "{:.2f}".format(v), color='black')
    plt.xlabel(column)
    plt.title(f"{column} v/s Formal Education")
    plt.savefig(f'./blog/{column}.png')
    return plt.show()
```

```python
plot_formaleducation(df, 'Salary')
```

```python
plot_formaleducation(df,'CareerSatisfaction')
```

```python
plot_formaleducation(df,'JobSatisfaction')
```
