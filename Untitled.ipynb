## Number of people per country

```python
def plot_country(df, filename):
    """
    This Function takes dataframe and filename as input and
    plots number of responders per country and
    saves the plot with name filename.png
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame on whose data is to be plotted
    filename : str
        Filename of image to be saved

    Returns
    -------
    function plt.show()
    """
    plt.figure(figsize=(20,10))
    ax = df['Country'].value_counts()[:20].plot(kind = 'bar', width = 0.8, color = '#9c9054') #plots the top 20 country
    for p in ax.patches:
        ax.annotate(format(p.get_height()),(p.get_x()+0.15, p.get_height()+10)) # writes the value on the top of the bar
    #set the labels
    plt.xlabel('Country') 
    plt.ylabel('Count')
    plt.title('Number of Responders per country (Top 20)')
    plt.savefig(f'./blog/{filename}.png')
    return plt.show()
    
plot_country(df, 'countywise')
```

## Number of students per country

```python
student = df[df['Professional']=='Student'] #create student dataframe.
plot_country(student,'studcount') #plot the student dataframe usign plot_country function
```

```python
#counter for languages
def language_counter(df):
    """
    counts the occurence languages
    parameter:
    -------------------
    df: dataframe
    which contains column 'HaveWorkedLanguage' 
    returns:
    -------------------
    languages: dictionary
    counts of different values
    """
    sd = df['HaveWorkedLanguage'].dropna() #drop nan tuples
    languages = {}
    for i in sd.values:
        for l in i.split(';'):
            languages[l.strip()] = languages[l.strip()] + 1 if l.strip() in languages else 1
    return languages
```

```python
# print(f"There are {len(languages)} Languages")
# plotting function for language plot
def show_languages(lang,filename):
    """
    This Function takes dictionary and filename as input and
    plots number of responders who use different languages and
    saves the plot with name filename.png
    Parameters
    ----------
    lang : Dictionary
        Dictionary containing the count of instances of different languages
    filename : str
        Filename of image to be saved

    Returns
    -------
    function plt.show()
    """
    lang = {k: v for k, v in sorted(lang.items(), key=lambda item: item[1])}
    plt.figure(figsize=(15,15))
    ax = pd.Series(lang).plot(kind='barh', width = 0.8, color = '#9c9054')
    for i,v in enumerate(lang.values()):
        ax.text(v + 3, i - 0.2, str(v), color='black')
    plt.xlabel("Count")
    plt.ylabel("Languages")
    plt.title("Number of Respondents v/s Language")
    plt.savefig(f'./blog/{filename}.png')
    return plt.show()
```

## Top Languages in Survey

```python
languages = language_counter(df)
show_languages(languages, 'alllang')
```

## Top Languages between students 

```python
languages = language_counter(student)
show_languages(languages,'studentlang')
```

```python
#formal education Plot
def plot_formaleducation(df, column):
    """
    This Function takes dataframe and filename as input and
    plots the average salary of people from different formal education background and
    saves the plot with name filename.png
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame on whose data is to be plotted
    filename : str
        Filename of image to be saved

    Returns
    -------
    function plt.show()
    """
    plt.figure(figsize=(20,10))
    dt = df.dropna(subset=[column]).groupby('FormalEducation').mean()[column]
    ax = dt.plot(kind = 'barh', width = 0.8, color = '#9c9054')
    for i,v in enumerate(dt):
        ax.text(v + v/200, i, "{:.2f}".format(v), color='black')
    plt.xlabel(column)
    plt.title(f"{column} v/s Formal Education")
    plt.savefig(f'./blog/{column}.png')
    return plt.show()
```

```python
plot_formaleducation(df, 'Salary') # Salary Plot
```

```python
plot_formaleducation(df,'CareerSatisfaction') #Career Satisfaction Plot
```
